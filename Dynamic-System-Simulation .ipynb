# Dynamic-System-Simulation 

There is a one amoeba in a pond. After every minute the amoeba may die, stay the same, 
split into two or split into three with equal probability. All its offspring, if it has any, will 
behave the same (and independent of other amoebas). What is the probability the 
amoeba population will die out?


```python
import random

def simulate_amoebea_efficient(max_minutes, num_simulations):

    print(f"\nSimulating {num_simulations} populations over {max_minutes} cycles")
    
    # Storing the incremental death probabilities
    max_minutes_popu = []
    death_probabilities =[]
    dead_population_count = 0

    for popu_num in range(1,num_simulations+1):
        population = 1

        for min in range(1,max_minutes+1):
            if population == 0: #If population is dead
                break

            next_population = 0
            for _ in range(population): # Simulate the fate of each amoeba in the population
                fate = random.choice(["die", "nothing", "split2", "split3"])
                if fate == "die":
                    continue
                elif fate == "nothing":
                    next_population += 1
                elif fate == "split2":
                    next_population += 2
                elif fate == "split3":
                    next_population += 3

            population = next_population

        if population == 0: # incrementing dead_population_count
            print(f'{popu_num}, ends at:{min}')
            dead_population_count += 1
        else:
          print(f'{popu_num}, survived')

        # Saving the minute reached in the current simulation
        max_minutes_popu.append(min)

        # Calculate the death probability incrementally
        death_probabilities.append(dead_population_count / popu_num)


    print(f'Simulating {num_simulations} populations over {max_minutes} cycles, death ratio of: {death_probabilities[-1]}')

    return max_minutes_popu,death_probabilities
```

```python
# Run the simulation
max_cycles_10_1000, probabilities_of_death_10_1000  = simulate_amoebea_efficient(max_minutes=10,num_simulations=1000)
```

```python
# Run the simulation
max_cycles_30_1000, probabilities_of_death_30_1000 = simulate_amoebea_efficient(max_minutes=30,num_simulations=1000)
```

```python
import matplotlib.pyplot as plt

# Plot the incremental death probabilities
plt.figure(figsize=(14, 7))
plt.plot(probabilities_of_death_30_1000, marker='o', linestyle='-', color='blue',label='30 Minutes Simulation')
plt.plot(probabilities_of_death_10_1000, marker='x', linestyle='-', color='red',label='10 Minutes Simulation')

# Set the title and labels
plt.title('Incremental Death Probabilities of Amoeba Populations')
plt.xlabel('Number of Tests')
plt.ylabel('Probability of Population Death')

plt.legend()
# Show a grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
